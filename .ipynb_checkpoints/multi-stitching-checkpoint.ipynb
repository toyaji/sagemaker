```python
# %load ./multiple_stitching.py
from imutils import paths
import numpy as np
import argparse
import imutils
import cv2
from tqdm.notebook import tqdm
import os
from moviepy.editor import ImageSequenceClip

"""
#construct the argument parser and parse the arguments
ap = argparse.ArgumentParser()
ap.add_argument("-r", "--right", type=str, required=True, help="path to input directory of right video to stitch")
ap.add_argument("-l", "--left", type=str, required=True, help="path to input directory of left to stitch")
ap.add_argument("-o", "--output", type=str, required=True, help="path to the output image")
ap.add_argument("-w", "--width", type=int, required=True, help="path to the output image")
args = vars(ap.parse_args())
"""

def stitching(args):

    #grap the paths to the input images and initialize our images list
    left_video = cv2.VideoCapture(args['left'])
    right_video = cv2.VideoCapture(args['right'])
    print('[INFO]: {} and {} loaded'.format(args['right'].split('/')[-1],
                                                    args['left'].split('/')[-1]))
    print('[INFO]: Video stitching starting....')

    # Get information about videos
    n_frames = min(int(left_video.get(cv2.CAP_PROP_FRAME_COUNT)), int(right_video.get(cv2.CAP_PROP_FRAME_COUNT)))
    fps = int(left_video.get(cv2.CAP_PROP_FPS))
    frames = []
    stitcher = cv2.Stitcher_create()
    
    # loop over the iamge paths, load each one, and add them to our images to sitch list
    for frame in tqdm(np.arange(n_frames)):
        ok, left = left_video.read()
        _, right = right_video.read()
    
        if ok:
            (status, stitched) = stitcher.stitch([left, right])

            if status == 0:
                resized = cv2.resize(stitched, (1000, 400), 0, 0, cv2.INTER_LINEAR)
                frames.append(resized)

            else: 
                print("[INFO] images stitching failed at {} th frame".format(str(frame)))

    return frames
```

```python
args = {'left':  "./data/left.mp4", 
'right': "./data/right.mp4",  
'output':"./data/output.mp4", }

frames = stitching(args)
```
